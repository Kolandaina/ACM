```python
## 实验一：手写数字5的特征分析
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
path = r'D:\中文\课程\机器学习\zip.train'
data = np.loadtxt(path)  # 7291*257, first column is number
# 筛选出数字5的图片
id5 = data[:, 0] == 5
data5 = data[id5,1:]
# 查看数字3图片数量
print("数字5的图片数量为:", data5.shape[0])
# 绘制一张数字5图片
plt.imshow(data5[555, :].reshape((16,16)))


# 创建PCA模型
pca = PCA(n_components=3)
# 使用pca拟合数据
pca.fit(data5)
# 获取降维后的数据矩阵
x = pca.transform(data5)
# 查看降维后的数据维度
print("降维后的维度为：",x.shape)
# 获取第一主成分方向的得分(投影值)排序
id1 = x[:,0].ravel().argsort()
# 绘制第一主成分方向得分(投影值)最大的五张图片、得分最低的五张图片
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(data5[id1[-i - 1]].reshape(16, 16))
for i in range(5):
    plt.subplot(2, 5, 5 + i + 1)
    plt.imshow(data5[id1[i]].reshape(16, 16))
plt.show()
# 获取第二主成分方向的得分排序
id2 = x[:, 1].ravel().argsort()


# 绘制第二主成分方向得分最大的五张图片、得分最低的五张图片
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(data5[id2[-i - 1]].reshape(16, 16))
for i in range(5):
    plt.subplot(2, 5, 5 + i + 1)
    plt.imshow(data5[id2[i]].reshape(16, 16))

plt.show()
# 获取第三主成分方向的得分排序
id3 = x[:, 2].ravel().argsort()

# 绘制第三主成分方向得分最大的五张图片、得分最低的五张图片
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(data5[id3[-i - 1]].reshape(16, 16))
for i in range(5):
    plt.subplot(2, 5, 5 + i + 1)
    plt.imshow(data5[id3[i]].reshape(16, 16))

# 显示图形
plt.show()
```
